# Capstone 2 Fitbit EDA

```python
# Importing needed packages

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importing cleaned data

FBdata = pd.read_csv('cleanedFBdata.csv', index_col='Unnamed: 0')
```

```python
FBdata.head()
```

```python
# list of all columns in our data

FBdata.keys()
```

```python
FBdata.shape
```

```python
# Checking the data types of our columns to verify that they are the correct type to plug in to our later models

FBdata.dtypes
```

```python
# Setting our date column to an actual date type

FBdata['Date'] = pd.to_datetime(FBdata['Date'])
```

```python
FBdata.dtypes
```

```python
# changing other data types in our data. Using a float type instead of interger because these columns are not categorical.

FBdata['Steps'] = FBdata['Steps'].astype(float)
FBdata['High_Activity_Min'] = FBdata['High_Activity_Min'].astype(float)
FBdata['Mod_Activity_Min'] = FBdata['Mod_Activity_Min'].astype(float)
FBdata['Light_Activity_Min'] = FBdata['Light_Activity_Min'].astype(float)
FBdata['Sed_Min'] = FBdata['Sed_Min'].astype(float)
FBdata['Calories'] = FBdata['Calories'].astype(float)
```

```python
# checking that our types are what we want

FBdata.dtypes
```

```python
# we will drop the date column and look at correlation of our column data

FBdataNT = FBdata.drop('Date', axis=1)
```

```python
FBdataNT.head()
```

```python
# This chart gives us an idea of which coulumns correlate and how closely they do. It may be required to drop a couple columns
# in the future for our regression model due to coliniarity such as the activity distance and  activity minutes columns. 


FBdataNT.corr()
```

```python
# Helping us to see what variables correlate with one another, we put them onto a heatmap so we can quickly see correlation

plt.figure(figsize=(20,10))
sns.heatmap(FBdataNT.corr(), annot=True)
```

```python
# I want to now group my data by participant ID and acquire the mean of each column to see if this changes our correlation and see if it may help in
# predicting results later

FBdata_by_par = FBdataNT.groupby(FBdataNT['Id']).mean()
```

```python
FBdata_by_par.head()
```

```python
# As we can see by the darker squares, this didn't show much more correlation among variables, but we also lost 
# quite alot of data by grouping each participants entries together

plt.figure(figsize=(20,10))
sns.heatmap(FBdata_by_par.corr(), annot=True)
```

```python
# The last 3 variables show very low correlation. Knowing they have some missing data, i wanna see how much again.
# Since we had 844 samples, these columns are missing almost half the data. For the moment we will drop these rows 
# with missing data, as i'm not sure how imputing half the data would affect our real data.

FBdataNT.isna().sum()
```

```python
FBdataNT.shape
```

```python
FBdata_DN = FBdataNT.dropna()
```

```python
# This only leaves us with 179 rows, or 21% of our original data. We will see how the sleep and HR columns correlate with other
# cloumns without the missing values affecting the results

FBdata_DN.shape
```

```python
FBdata_DN.corr()
```

```python
# Without the missing data, the sleep and HR columns show much more correlation, but not a great deal. 

plt.figure(figsize=(20,10))
sns.heatmap(FBdata_DN.corr(), annot=True)
```

```python
# We will look at the other columns that we have complete data on

FBdata_clean = FBdataNT.drop(['Heart_Rate', 'Time_Asleep', 'Time_In_Bed'], axis=1)
```

```python
FBdata_clean.head()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(FBdata_clean.corr(), annot=True)
```

```python
# we can use a pairplot to get another idea of correlation and the view of our data distribution

sns.pairplot(FBdata_clean)
```

```python
# Sedinatry activity distance is mostly zero as we would expect. I think we can drop this variable

FBdata_clean['Sed_Activity_Dist'].describe()
```

```python
FBdata_clean = FBdata_clean.drop('Sed_Activity_Dist', axis=1)
```

```python
FBdata_clean['High_Activity_Min'].hist()
```

```python
FBdata_clean['Mod_Activity_Min'].hist()
```

```python
FBdata_clean[FBdata_clean['Calories'] < 1000]
```

```python
# Distance to calories correlation looks to be a huge predictor variable for us.

sns.regplot('Distance','Calories', data=FBdata_clean)
```

```python
# lets take a look at one participant's data

user_data1 = FBdata[FBdata['Id'] == 1503960366]
```

```python
user_data1.head()
```

```python
# amount of data from this user

user_data1.shape
```

```python
# Lets build a graph to see how correlated our Distance and Calories data look for this user.

fig,ax = plt.subplots()
ax.plot('Date', 'Calories', data=user_data1, alpha=0.5)
ax.set_xlabel('Date')
ax.tick_params(axis='x', labelrotation=45)
ax.set_ylabel('calories')
ax2 = ax.twinx()
ax2.plot('Date', 'Distance', data=user_data1, color='red', alpha=0.5, linestyle='--')
ax2.set_ylabel('distance')
ax.legend()
ax
```
